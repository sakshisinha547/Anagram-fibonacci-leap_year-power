```python
def anagram(string1 :str , string2:str):
    
    if not isinstance(string1,str) or not isinstance(string2,str):
        raise TypeError("input must be string")
        
    string1 = string1.replace(" ","").lower() 
    string2 = string2.replace(" ","").lower()      
    
    if len(string1) != len(string2):
        print(False)

    char_count = {}
    for char in string1:
        if char in char_count:
            char_count[char]+=1
        else:
            char_count[char] =1
            print(char_count)
    for char in string2:
        if char not in char_count:
            print(False)
            break
        char_count[char] -=1
        if char_count[char] <0:
            return False
            break
    else:
        return True        
if __name__ == "__main__":
    try:
        string1 = input("enter your string")
        string2 = input("enter your string")
        result = anagram(string1,string2)
        print(result)
             
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print(e)

        
```

```python
def leap_year(year:int):
    # type check
    if not isinstance(year,int):
        raise TypeError("input must be an integer")
    if year <1582 :
        raise ValueError("year must later on 1582")
    if year >1582 and(year %4 == 0 and year % 100 != 0 ) or (year % 400 == 0):
        return True
    else:
        return False   
#____main excuetion_____
if __name__ == "__main__":
    #user_input
    year = 2000
    try:
        # calling a function
        result = leap_year(year)
        print(result)
    except (TypeError,ValueError) as e:
        print(e)
    except Exception as e:
        print(e)
```

```python
def find_power(base_value:int , power:int):
    # type check
    if not isinstance(base_value,int) or not isinstance(power ,int):
        raise TypeError("input must be an integer")
    # check if the value is negetive    
    if power <0:
        raise ValueError("power value must be positive")
    result = 1
    for i in range(power):
        result = result *base_value
    return result 
#________main excuetion___
if __name__ == "__main__":
    
    try:
        # taking inout form user
        base_value = int(input("enter a number:"))
        power = int(input("enter a number:")) 
        # calling a function
        result = find_power(base_value,power)
        print(result)
    except (TypeError,ValueError) as e:
         print(e)
    except Exception as e:
         print(e)
     
```
